### **1. Line plot**  
Use to do bivariate analysis on two columns  
both column could be numerical OR categorical OR mixed.

and y-axis depends on x, therefor the independet variable will always go to x.  

Example - consider two columns, Num_Students and Attendance%, here Attendance% depends on Num_Students, therfore Attendance% will go to y-axis and Num_Students to x-axis.

```python
#data
price= [48000,56000,52000,41000,43000]
year= [2017,2018,2019,2020,2021]

plt.plot(year,price, label= 'Iphone 12 Price')
plt.legend()
plt.xlabel('Year')
plt.ylabel('Price')
```

### **2. Scatter plot**  
A scatter plot is used to show the correlation between two numerical columns. Its the base version of line plot except the dots are not connected with a line.  

- A scatter plot can be used for-  
bivariate analysis.  
numerical vs numerical.  
show correlation between two columns.  

_**NOTE**_ - here any column can be plotted in both axis, means there is no concept of dependent and independent (variable most of the time).

```python
batter = pd.read_csv('/content/drive/MyDrive/DataScinece/Datasets/batter.csv')
bat = batter.head(600)
plt.scatter(bat['avg'], bat['strike_rate'], marker= 'o')
```

### **3. Bar plot**  
Bar chart is use to represent the quantity of category across the data.  
- Bar char uses-
one catogorical column in x-axis and y-axis represent an aggregate value like- quantity etc.  
Bar chart can be used for both univariate and bivariate column analysis.

```python
child = [11,17,9,22,14]
color = ['green', 'red', 'black', 'orange', 'yellow']
plt.bar(color, child)
```

```python
#horizontal bar
plt.barh(color, child)
```

```python
#plotting multiple bars on one graph
df = pd.read_csv('/content/drive/MyDrive/DataScinece/Datasets/batsman_season_record.csv')
df
```

```python
plt.bar(np.arange(df.shape[0])+0.33,df['2015'], width= 0.33, label= '2015')
plt.bar(np.arange(df.shape[0]),df['2016'], width= 0.33, label= '2016')
plt.bar(np.arange(df.shape[0])-0.33,df['2017'], width= 0.33, label= '2017')

plt.xticks(np.arange(df.shape[0]), df['batsman'], rotation=45)
plt.ylabel('Runs')
plt.legend()
```

```python
#stacked bar chart
plt.bar(df['batsman'], df['2017'], label= '2017')
plt.bar(df['batsman'], df['2016'], bottom= df['2017'],label= '2016')
plt.bar(df['batsman'], df['2015'], bottom= df['2016']+df['2017'],label= '2015')
plt.legend()
plt.ylabel('runs_contribution')
```

### **3. Pie Chart**  
A Pie chart is used to show the part of a catogory out of whole 100%.  
- used in catogory columns.  
- can be use to do univariate/bivariate analysis.  
- both on numerical and catogory columns.

```python
marks = [55,95,40,70]
subject = ['Maths', 'English', 'Physics', 'Chemistry']
plt.pie(marks, labels=subject, autopct='%0.1f%%', explode=[0.1,0,0,0], shadow= True)
plt.title('Marks Distribution', fontweight = 'bold', fontsize= 12)
```

```python
batter= pd.read_csv('/content/drive/MyDrive/DataScinece/Datasets/batter.csv')
batter
```

```python
plt.figure(figsize=(10,5))

sample = batter.sample(20, random_state=42)

#variables
sr_median = sample['strike_rate'].median()
avg_median = sample['avg'].median()

#plotting
plt.scatter(sample['avg'], sample['strike_rate'], s= sample['runs'], alpha= 0.7)
plt.axhline(sr_median, color= 'green', linestyle = '--',label=f'Median SR:{sr_median:.1f}')
plt.axvline(avg_median, color= 'red', linestyle = '--', label= f'Median Avg:{avg_median:.1f}')
plt.xlabel('Avarage')
plt.ylabel('StrikeRate')
plt.title('Avg v/s StrikeRate', fontdict={'size':15, 'fontweight':'bold'})
plt.legend(loc = 'lower right')
plt.tight_layout()  #automatically adjusts the elements
plt.grid(alpha= 0.3)  #for grid opacity

#labeling names
for i in range(sample.shape[0]):
    plt.text(sample['avg'].values[i], sample['strike_rate'].values[i], sample['batter'].values[i], fontdict={'size':8, 'color':'black'})
```

### **2. Subplots**  
- Subplots are used to plot multiple graph in a single figure.  
- It good for univariate analysis where each column need a saperate graph
so rather than building saperate.


```python
top = batter.head(10)
top

fig, ax = plt.subplots(nrows= 2, ncols= 2, figsize=(10,4))
ax[0,0].hist(top['runs'], color = 'green')
ax[0,0].set_title('runs')
ax[0,0].axvline(top['runs'].median(), linestyle= '--', color= 'red')
ax[0,0].axvline(top['runs'].mean(), linestyle= 'solid', color= 'green')

ax[1,0].hist(top['avg'])
ax[1,0].set_title('avg')

ax[0,1].hist(top['strike_rate'])
ax[0,1].set_title('strike_rate')


fig.tight_layout()
```

### **3. Grid map**

```python
delivery = pd.read_csv('/content/drive/MyDrive/DataScinece/Datasets/deliveries.csv')
delivery.head()
```

```python
temp = delivery[(delivery['ball'].isin([1,2,3,4,5,6])) & (delivery['batsman_runs'] ==6)]
```

```python
temp2 = temp.pivot_table(index= 'over', columns= 'ball', values='batsman_runs', aggfunc='count')
temp2.head()
```

```python
plt.figure(figsize= (10,5))
plt.imshow(temp2)
plt.title('Heatmap', fontweight= 'bold', fontsize= 12)
plt.colorbar()
plt.tight_layout()
```
